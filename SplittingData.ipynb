```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
#Assigning data and target to seperate variables
dtf = pd.read_excel("../State Population Data.xlsx", sheet_name='Final Sheet1')
population_target = pd.DataFrame(dtf['index_nsa'])

dtf = dtf.drop('index_nsa', axis=1)
dtf = dtf.drop('index_sa', axis=1)
dtf
# population_target
```

```python
# population_target
#Splitting the Data 70-30
X_train, X_test, y_train, y_test = train_test_split(
    dtf, population_target, train_size= 0.70, random_state=23)
X_train.head()
```

```python
y_train.head()
```

```python
X_train.to_csv('X_train.csv')
X_test.to_csv('X_test.csv') 
y_train.to_csv('y_train.csv')
y_test.to_csv('y_test.csv')
```

```python
#Splitting data 70-30 by state
X_train_byState = pd.DataFrame()
X_test_byState = pd.DataFrame() 
y_train_byState = pd.DataFrame()
y_test_byState = pd.DataFrame()

num_of_states = int(len(dtf.index)/10)

for i in range(num_of_states):
    #Retrieving all the rows for an individual state
    df_state = dtf[i*10 : (i*10)+10]
    df_state_target = population_target[i*10 : (i*10)+10]

    #Splitting state data 70-30
    X_train, X_test, y_train, y_test = train_test_split(
        df_state, df_state_target, train_size= 0.70, random_state=23)

    X_train_byState = X_train_byState.append(X_train)
    X_test_byState = X_test_byState.append(X_test) 
    y_train_byState = y_train_byState.append(y_train)
    y_test_byState = y_test_byState.append(y_test)

# X_train_byState
```

```python
X_train_byState.to_csv('X_train_byState.csv')
X_test_byState.to_csv('X_test_byState.csv') 
y_train_byState.to_csv('y_train_byState.csv')
y_test_byState.to_csv('y_test_byState.csv')
```
